```
import pandas as pd
import numpy as np
import nltk
import csv
import sys

csv.field_size_limit(sys.maxsize)

paper_id,abstract,text,authors,title,journal,summary=[],[],[],[],[],[],[]



with open('initial_coviddata.csv','r') as f:
  df=csv.reader(f)

  for row in df:
    paper_id.append(row[1])
    abstract.append(row[2])
    text.append(row[3])
    authors.append(row[4])
    title.append(row[5])
    journal.append(row[6])
    summary.append(row[7])

data_dict=dict(zip(['paper_id', 'abstract', 'body_text', 'authors', 'title', 'journal', 'abstract_summary'],[paper_id,abstract,text,authors,title,journal,summary]))
pd.DataFrame(data_dict)
    
```
